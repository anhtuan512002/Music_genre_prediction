```python
#hảm lấy thông tin page
def get_page_content(url):
  page=requests.get(url)
  return bs4.BeautifulSoup(page.text,"html.parser")
```

```python
# preprocesing
import re
def Preprocessing(string):
  # Xóa các ký tự kéo dài: vd: đẹpppppp
  string = re.sub(r'([A-Z])\1+', lambda m: m.group(1).upper(), string, flags=re.I)
  string = string.lower()

  # link
  string = re.sub('<.*?', '', string).strip()
  string = re.sub('(\s)+', r'\1', string)

  # xóa ký tự đặc biệt
  string = re.sub(r"[-()\\\"#/@;:<>{}`+=~|.!?,%/]", "", string) # có phẩy
  string = re.sub('\n', ' ', string)
  return string
```

```python
#hàm pre 1
def pre1(X):
  X_1=[]
  for i in range(len(X)):
    a=Preprocessing(X[i][0])
    b=Preprocessing(X[i][1])
    c=Preprocessing(X[i][2])
    ds=[]
    ds.append(a)
    ds.append(b)
    ds.append(c)
    X_1.append(ds)
  return X_1
```

```python
#hàm pre 2
def pre2(X):
  X_1=[]
  for i in range(len(X)):
    a=Preprocessing(X[i][0])
    X_1.append(a)
  return X_1
```

```python
#hàm pre 3
def pre3(X):
  X_1=[]
  for i in range(len(X)):
    a=Preprocessing(X[i][0])
    b=Preprocessing(X[i][1])
    ds=[]
    ds.append(a)
    ds.append(b)
    X_1.append(ds)
  return X_1
```

```python
#hàm pre 4
def pre4(X):
  X_1=[]
  for i in range(len(X)):
    c=Preprocessing(X[i][2])
    X_1.append(c)
  return X_1
```

```python
#hàm xử lí
def TfldfVector(X_train, X_test):
    tfidf = TfidfVectorizer()
    X_train_encode = tfidf.fit_transform(X_train)
    X_test_encode = tfidf.transform(X_test)
    return X_train_encode, X_test_encode
```
